```python
import numpy as np
from skimage.feature import corner_harris,corner_peaks
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.exposure import equalize_hist
def show_corners(corners,image):
    fig=plt.figure()
    plt.gray()
    image=color.gray2rgb(image)
    plt.imshow(image)
    y_corner,x_corner=zip(*corners)
    plt.plot(x_corner,y_corner,'or')
    plt.xlim(0,image.shape[1])
    plt.ylim(image.shape[0],0)
    fig.set_size_inches(np.array(fig.get_size_inches())*1.5)
    plt.show()
    
    
yadav=io.imread('downloads/yadav1.jpg')
yadav=equalize_hist(rgb2gray(yadav))
corners=corner_peaks(corner_harris(yadav),min_distance=2)
show_corners(corners,yadav)
    
    

    

```

```python
# draw line circle rectangle on image 
from skimage import io
from skimage import color
from skimage import draw
img=io.imread('downloads/yadav1.jpg')
x,y=draw.line(0,0,1000,1000) #it draw line from 0,0 to 1000,1000 
x1,y1=draw.circle(300,300,100)
img[x1,y1]=255
#img[x,y]=255 #it change the value of pixels from line to black
#0 for black with that value you can decide the color of line 255 for white
io.imshow(img)

```

```python
#gamma correction it is show the luminess with pixel values
from skimage import io
from skimage import exposure
from pylab import *
img=io.imread('downloads/yadav1.jpg')
gamma_corrected1=exposure.adjust_gamma(img,1)
gamma_corrected2=exposure.adjust_gamma(img,5)
figure(0)
io.imshow(gamma_corrected1)
figure(1)
io.imshow(gamma_corrected2)
```

```python
# determine structure similarity 
"""this is use to find the similarity between the pixel value 
  value closer to 0 mean they have less similarity 
  value closer to 1 mean they have high similarity

"""
from skimage import io
from skimage.measure import compare_ssim as ssim
img_original=io.imread('downloads/yadav1.jpg')
img_modified=io.imread('downloads/yadav1.jpg')
ssim_original=ssim(img_original,img_original,data_range=img_original.max()-img_original.min(),multichannel=True)
ssim_different=ssim(img_original,img_modified,data_rang=img_original.max()-img_modified.min(),multichannel=True)
print(ssim_original,ssim_different)















```
